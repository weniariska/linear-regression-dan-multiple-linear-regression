<h3>IMPORT CLEAN DATA</h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# path of data 
path = 'clean_df_automobile.csv'
df = pd.read_csv(path)
df.head()
```

<h3>LINEAR REGRESSION</h3>


Y hat = a + bX

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()                               # buat objek linear regression
lm.fit(df[['highway-L/100km']], df[['price']])        # membuat fungsi linear dengan highway-L/100km sebagai prediktor dan "price" sebagai response variable
Yhat=lm.predict(df[['highway-L/100km']])              # Y hat = a + bX

print("contoh 1")
print("==================================================")
print("Prediktor(X) : highway-L/100km | Response(Y) : price")
print("==================================================")
print("5 prediksi harga teratas : \n", Yhat[0:5])
print("\nIntercept     (a) : ", lm.intercept_)
print("Gradien/Slope (b) : ", lm.coef_)
print("model linear -> price =", lm.intercept_[0] ,"+", lm.coef_[0][0] ,"x highway-L/100km")
print("==================================================")


# engine-size dan price
lm1 = LinearRegression()
lm1.fit(df[['engine-size']], df[['price']])
print("\n\ncontoh 2")
print("==================================================")
print("Prediktor (X) : engine-size | Response (Y) : price")
print("==================================================")
print("Intercept     : ", lm1.intercept_)
print("Gradien/Slope : ", lm1.coef_)
print("model linear -> price =", lm1.intercept_[0] ,"+", lm1.coef_[0][0] ,"x engine-size")
print("==================================================")
```

<h3>MULTIPLE LINEAR REGRESSION</h3>


Y hat = a + b1X1 + b2X2 + ... + bnXn

```python
lm2 = LinearRegression()
lm2.fit(df[['horsepower', 'curb-weight', 'engine-size', 'highway-L/100km']], df['price'])
print("==================================================")
print("            MULTIPLE LINEAR REGRESSION")
print("==================================================")
print("Intercept     : ", lm2.intercept_)
print("Gradien/Slope : ", lm2.coef_)
print("model linear -> price =", lm2.intercept_ ,"+", lm2.coef_[0] ,"x horsepower +", lm2.intercept_ ,"+", lm2.coef_[1], "x curb-weight", lm2.intercept_ ,"+", lm2.coef_[2], "x highway-L/100km")
print("==================================================")
```

```python
# length dan price
lm3 = LinearRegression()
lm3.fit(df[['length']], df[['price']])
print("Intercept     : ", lm3.intercept_)
print("Gradien/Slope : ", lm3.coef_)
```
